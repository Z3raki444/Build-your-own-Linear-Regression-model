```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
# x represents the independant variable (feature)

y = np.array([2,4,6,8,10])
# y represents the dependent variable (target/output)
```

```python
print(x.shape)
# shows number of samples in x

print(y.shape)
# shows number of samples in y
```

```python
x = x.reshape(-1,1)
# Converts x from shape (5,) to (5, 1)
# -1 means "figure out rows automatically"
# 1 means one feature
```

```python
m = 0
# m is the slope (weight)
```

```python
b = 0
# b is the intercept (bias)
```

```python
y_pred = x*m +b
#This is the linear regression formula: y = mX + b
# Currently predictions will be all zeros
```

```python
error = y - y_pred
# Measures how far predictions are from true values
```

```python
mse = np.mean(error ** 2)
# Squares the errors and takes their average
# This is the loss function we want to minimize
```

```python
print(mse)
# Higher value = worse model
```

```python
learning_rate = 0.01
# Learning rate controls how big each update step is
# Too big → model overshoots
# Too small → learning is very slow
```

```python
epochs = 1000
#one epoch  = one full update of m and b
# more epochs = more learning
```

```python
for i in range(epochs):
    # Loops reuns 1000 times to improve m and b gradually
    y_pred = m * x + b
    # Predict y using current values of m and b
```

```python
    dm = (-2 / len(x)) * np.sum(x * (y - y_pred))
    # Gradient of loss w.r.t slope (m)
    # Tells us how m should change

    db = (-2 / len(x)) * np.sum(y - y_pred)
    # Gradient of loss w.r.t intercept (b)
    # Tells us how b should change
```

```python
    m = m - learning_rate * dm
    # Move m in the direction that reduces error

    b = b - learning_rate * db
    # Move b in the direction that reduces error
```

```python
    if i % 100 == 0:
        loss = np.mean((y - y_pred) ** 2)
        print(f"Epoch {i}: m={m:.2f}, b={b:.2f}, loss={loss:.2f}")
```

```python
print("Final slope (m):", m)
print("Final intercept (b):", b)

```
